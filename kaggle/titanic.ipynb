```python
import pandas as pd

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
train.info() #pclass, sex, age, sibsp, parch, Fare
```

```python
test
```

```python
titanic_train_data = train[['Pclass', 'Sex', 'Age','Fare']]
titanic_test_data = test[['Pclass', 'Sex', 'Age', 'Fare']] #컬럼 지정

titanic_train_target = train['Survived']
```

```python
titanic_train_data['Age'] = titanic_train_data['Age'].fillna(titanic_train_data['Age'].mean())
titanic_test_data['Age'] = titanic_test_data['Age'].fillna(titanic_test_data['Age'].mean())

titanic_test_data['Fare'] = titanic_test_data['Fare'].fillna(titanic_test_data['Fare'].mean())
#결측값 대치(age, fare를 평균으로)
```

```python
titanic_train_data['Sex'] = titanic_train_data['Sex'].replace({'male': 1, 'female': 2})
titanic_test_data['Sex'] = titanic_test_data['Sex'].replace({'male': 1, 'female': 2})
```

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier #모듈 불러오기

rf = RandomForestClassifier(n_jobs=-1)
scores = cross_validate(
    rf, 
    titanic_train_data, 
    titanic_train_target, 
    return_train_score=True, 
    n_jobs=1
)
print(scores)

rf.fit(titanic_train_data, titanic_train_target)

predictions = rf.predict(titanic_test_data)

#예측데이터 프레임
submission2 = pd.DataFrame({'PassengerId': range(892, 1310), 'Survived': predictions})

# CSV 저장
submission2.to_csv('submission2.csv', index=False)
```

```python

rf = RandomForestClassifier(oob_score=True, n_jobs=1)
rf.fit(titanic_train_data, titanic_train_target)
rf.oob_score_ #자체적인 validation score를 보여주는 중입니다,
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier

hgb = HistGradientBoostingClassifier()
scores = cross_validate(
    hgb, titanic_train_data, titanic_train_target,
    return_train_score=True, n_jobs=-1)

print(scores)
```

```python
from sklearn.inspection import permutation_importance

hgb.fit(titanic_train_data, titanic_train_target)
scores = permutation_importance(hgb, titanic_train_data, titanic_train_target)
print(scores.importances_mean)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

dt = DecisionTreeClassifier(min_impurity_decrease=0.001)

params = {
    'min_impurity_decrease' : [0.001, 0.0002, 0.0003, 0.0004]
}

gs = GridSearchCV(dt, params, n_jobs=-1)
gs.fit(titanic_train_data, titanic_train_target)
```

```python
gs.best_estimator_
```

```python
dt = DecisionTreeClassifier(min_impurity_decrease=0.001)
dt.fit(titanic_train_data, titanic_train_target)
```

```python
scores = cross_validate(dt, titanic_train_data, titanic_train_target,return_train_score=True, n_jobs=-1)
print(scores)
```

```python
predictions = dt.predict(titanic_test_data)

#예측데이터 프레임
submission = pd.DataFrame({'PassengerId': range(892, 1310), 'Survived': predictions})

# CSV 저장
submission.to_csv('submission.csv', index=False)
```
